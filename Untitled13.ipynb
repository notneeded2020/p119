```python
import pandas as pd
import csv 

col_names = ["psngrid","pclass","sex","age","sibsp","parch","survived"]

df = pd.read_csv("titanic.csv", names = col_names).iloc[1:]
print(df.head())
```

```python
features = ["psngrid","pclass","sex","age","sibsp","parch"]
X = df[features]
Y = df.survived
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics

X_train,X_test,Y_train, Y_test = train_test_split(X,Y,test_size=0.3,random_state=1)
clf = DecisionTreeClassifier(max_depth = 4)
clf = clf.fit(X_train,Y_train)
Y_pred = clf.predict(X_test)
print("accuracy of given data: ", metrics.accuracy_score(Y_test,Y_pred))
```

```python
from sklearn.tree import export_graphviz
from IPython.display import Image 
import pydotplus
from six import StringIO

dot_data = StringIO()

export_graphviz(clf,
                out_file=dot_data,
                filled=True,
                rounded=True,
                special_characters=True,
                feature_names = features,
                class_names = ["0","1"])

print(dot_data.getvalue())

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png("titanic.png")
Image(graph.create_png())
```
